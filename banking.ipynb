# Preporcessing

```python
df = df.drop('day_month',axis =1 )
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder  , PowerTransformer
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

num = X.select_dtypes(include = 'number')
cat = X.select_dtypes(exclude = 'number')
```

```python
num.head()
```

```python
trf = PowerTransformer(standardize= True)
num_new = trf.fit_transform(num)
num_new = pd.DataFrame(num_new,columns = num.columns,index = num.index)
num_new.head()
```

```python
cat.head()
```

```python
ohe = OneHotEncoder(sparse_output = False ,dtype = np.int32)
cat_new = ohe.fit_transform(cat)
cat_new = pd.DataFrame(cat_new ,columns =ohe.get_feature_names_out() ,index = cat.index)
cat_new.head()
```

```python
label = LabelEncoder()
y = label.fit_transform(y)
y
```

```python
X_new = pd.concat([num_new,cat_new],axis = 1)
X_new.head()
```

```python
X_train ,X_test ,y_train ,y_test = train_test_split(X_new ,y ,test_size =0.25 ,shuffle =True)
print(X_train.shape ,X_test.shape)
```

# Model

```python
!pip install catboost
!pip install optuna
import catboost
from catboost import CatBoostClassifier
from sklearn.metrics import confusion_matrix ,accuracy_score ,classification_report
```

```python
import optuna
def objective(trial):

    params = {
        "iterations": trial.suggest_int("iterations", 300, 1000),
        "depth": trial.suggest_int("depth", 4, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1, 10),
        "bagging_temperature": trial.suggest_float("bagging_temperature", 0, 1),
        "random_strength": trial.suggest_float("random_strength", 0, 1),
        "loss_function": "Logloss",
        "eval_metric": "AUC",
        "verbose": 0,
        "random_seed": 42
    }

    model = CatBoostClassifier(**params)

    model.fit(
        X_train, y_train,
        eval_set=(X_test, y_test),
        early_stopping_rounds=50,
        verbose=False
    )

    preds = model.predict(X_test)
    score = accuracy_score(y_test, preds)

    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)
```

```python
print("Best score:", study.best_value)
print("Best parameters:", study.best_params)
```

```python
clf2 = CatBoostClassifier(iterations =329 ,depth =9 ,learning_rate = 0.0579 ,l2_leaf_reg =  2.53 ,bagging_temperature =0.208141)
clf2.fit(X_train ,y_train)
```

```python
print(accuracy_score(y_test ,clf2.predict(X_test)))
print(accuracy_score(y_train ,clf2.predict(X_train)))
```

```python
confusion_matrix(y_test ,clf2.predict(X_test))
```

```python
print(classification_report(y_test ,clf2.predict(X_test)))
```
